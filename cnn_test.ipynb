```python
ids = np.load('idsMatrix.npy')
ids.shape
```

```python
x = tf.placeholder(tf.float32, [None, 61])
```

```python
W = tf.Variable(tf.zeros([61, 2]))
b = tf.Variable(tf.zeros([2]))
```

```python
y = tf.nn.softmax(tf.matmul(x, W) + b)
```

```python
y_ = tf.placeholder(tf.float32, [None, 2])
```

```python
cross_entropy = tf.reduce_mean(-tf.reduce_sum(y_ * tf.log(y), reduction_indices=[1]))
```

```python
train_step = tf.train.GradientDescentOptimizer(0.5).minimize(cross_entropy)
```

```python
sess = tf.InteractiveSession()
tf.global_variables_initializer().run()
```

```python
max_seq_length = 61
batchSize = 100
lstmUnits = 64
numClasses = 2 # lib or con
iterations = 100000
numDimensions = 300 #Dimensions for each word vector
```

```python
# helper functions for training the network 
# [1,0] is lib
# [0,1] is con

from random import randint

num_libs = len(lib)
num_cons = len(con)
merged_list = lib + con

def getTrainBatch():
    labels = []
    arr = np.zeros([batchSize, max_seq_length]) #shape: (100, 61)
    # training batch is half lib and half con
    for i in range(batchSize):
        if (i % 2 == 0): 
            num = randint(1,num_libs-500)
            labels.append([1,0])
        else:
            num = randint(num_libs+500,num_libs+num_cons)
            labels.append([0,1])
        arr[i] = ids[num-1:num]
    return arr, labels

# 1000 test sentences
def getTestBatch():
    labels = []
    arr = np.zeros([batchSize, max_seq_length])
    for i in range(batchSize):
        num = randint(num_libs-500,num_libs+500)
        if (num <= num_libs):
            labels.append([1,0])
        else:
            labels.append([0,1])
        arr[i] = ids[num-1:num]
    return arr, labels
```

```python
for _ in range(1000):
  nextBatch, nextBatchLabels = getTrainBatch();
  sess.run(train_step, feed_dict={x: nextBatch, y_: nextBatchLabels})
```

```python
correct_prediction = tf.equal(tf.argmax(y,1), tf.argmax(y_,1))
```

```python
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

```python
sum = 0
for i in range(1000):
    nextBatch, nextBatchLabels = getTestBatch();
    acc = sess.run(accuracy, feed_dict={x: nextBatch, y_: nextBatchLabels}) * 100
    sum += acc
    print("Accuracy for this batch:", acc)

print sum
print sum/1000.0
    
```
